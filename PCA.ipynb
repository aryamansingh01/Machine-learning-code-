```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

```

```python
data = np.array([
    [-0.5, -0.2, 0.1],
    [0.8, 1.2, 1.5],
    [-1.1, -0.8, -0.9],
    [0.3, 0.5, 0.6]
    
])


scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)


pca = PCA(n_components=2)


data_pca = pca.fit_transform(data_scaled)


```

```python
print("Transformed data:\n", data_pca)


print("Explained variance ratio:", pca.explained_variance_ratio_)


cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
print("Cumulative explained variance:", cumulative_variance)





```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data_pca[:, 0], data_pca[:, 1], alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('/Users/aryamansingh/Desktop/CUB course/Machine learning /project/dataset/state_year_totals_electrci.csv')
df_pivot = df.pivot(index='State', columns='Year', values='Total Vehicles')
df_pivot.fillna(0, inplace=True)
```

```python
scaler = StandardScaler()
df_pivot_scaled = scaler.fit_transform(df_pivot)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_pivot_scaled)

df_pca = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2'])
plt.scatter(df_pca['PC1'], df_pca['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Vehicle Sales Data')
plt.show()
```

```python
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```
